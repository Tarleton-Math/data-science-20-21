To do data science well, you need be ALL of:
- Math/Stat problem solving & critical thinking
- Computer Science code & algorithm development
- Domain knowledge in the area you are applying data science

Why do you need ALL of those things:
- A data scientist without math/stats is prone tell lies.  Code will always give results (once debugged), but are they correct?  Models often look great when applied to the data we built them from but *fail spectacularly* when used on new data?  This "generalization performance" is the real objective, and it can be embarrasingly poor if we fail to apply healthy statistical skepticsim regarding issues like sampling bias, missing data, hidden/confounding variables, data anomalies, correlation vs causation, etc.

- A data scientist without computer science skills struggles with practical and implementation obstacles such as data cleaning/pre-processing, writing efficient and readable code, utilizing computational resources, training & tuning models, finding "good enough" solutions, etc.

- A data scientist without domain knowledge is shooting in the dark.  They are prone to really silly errors that would never be made by anyone who knows the business/field.  They waste lots of time and feel frustrated.  And even if they do somehow make a model that "works", there's a good chance it won't ever get used because it didn't answer a question the boss cares about.

# Basic Terminology

- The *domain* is the system we are investigating
- A *sample / observation* is an example / element in the domain, typically represented as a row in the dataset
- A *variable* is a property of interest in the domain, typically represented as a column in the dataset

- Variable Types
    - *Qualitative / Categorical* - a variable with a discrete set of values
        - *Ordinal* - a categorical variable where the values have a natural order
            - ex: "class year" has values Freshman / Sophmore / Junior / Senior
        - *Nominal* - a categorical variable where the values do not have a natural order
            - ex: "party"  has values Republican / Democrat / Libertarian / Green / Independent
    - *Quantitative / Continuous* - a variable with a continuouset of values
        - ex: weight has values [0, ∞)

        

- Supervision
    - [*Supervised learner*](https://en.wikipedia.org/wiki/Supervised_learning)
        - distinguishes two roles for variables
            - *Feature / Input / Predictor / Independent* - a variable that goes into the model & it uses to make predictions
            - *Target / Output / Dependent / Response* - a variable the model is trying to predict
        - has access to the "true" values of both features AND targets for all observations in the datset
        - attempts to predict the targets using the features in a way that will generalize to not-yet-observed samples.
            - If all target variables are quantitative, model called a *regressor*
                - Ex: predict corn yield for a farmer's field based on the rain, temperature, soil Nitrogen contenet, and seed type 
            - If all target variables are categorical, model called a *classifier*
                - Ex: classify income tax returns as "normal" or "anomalous" based on adjusted gross income, tax liability, and itemized deductions
            - Hard to handle mixed-type target variables
    - [*Unsupervised learner*](https://en.wikipedia.org/wiki/Unsupervised_learning)
        - does not distinguish variables as features / targets - treats them all (basically) the same

    - [*Semi-supervised learner*](https://en.wikipedia.org/wiki/Semi-supervised_learning)
        - alternates between supervised and unsupervised
        - Typical case: Lots of observations of features, but only a few obervations have been evaluated wrt the targets.
            - ex: lots of income tax returns, but only a few actually get audited.  Moreover, there is significant sampling bias toward anomalous class because audits are usually done if something "looks weird" (not simple random sample).
            - Supervised learner requires more unbiased samples of target variables, but this can be expensive to produce
            - Model attempts to guide review process to be a efficient as possible (greatest increase in data quality per sample evaluated)
    - [*Reinforcement learner*](https://en.wikipedia.org/wiki/Reinforcement_learning)
        - "Agent" operating in a environment subject to a reward function or opponent or external force, etc
        - You must watch this [OpenAI video](https://www.youtube.com/watch?v=kopoLzvh5jY).  It is the most delightful thing I've watched recently.


- Parameter vs Hyperparameter - Quantities used by a model that are not in (or derived directly from) the dataset
    - A *parameter* has a well-defined value that "best-fits" the training set that can be algorithmically found by a computer
        - *Training* - finding values for parameters the best fit the training set
    - A *hyperparameter* does not have a well-defined best value.
        - There is typically a trade-off between competing (un)desirable model characteristics
        - There are algorithms to help with this choice (grid search, gradient descent, genetic algorithms, etc), but the responsibility ultimately falls to the data scientist 
        - *Tuning / model selection* - finding values for hyperparamters with the optimal balance of all (un)desirable model characteristcs
    - Example
        - one continuous feature x
        - one continuous target y
        - Make scatter plot
        - Try to fit a curve (linear regression)
            - Fit a line: $y= \beta_0 + \beta_1 x$
            - Fit a quadratic: $y= \beta_0 + \beta_1 x + \beta_2 x^2$
            - Fit a cubic: $y= \beta_0 + \beta_1 x + \beta_2 x^2+ \beta_3 x^3$
            - ...
        - Hyperparameter: polynomial degree
            - No clear cut best choice
            - Higher degree ... 
                - Always fits training data better
                - Risks overfitting training data causing poor performance on validations & holdout sets
                - Tuning - Typically, try many different degrees and look at performance/complexity trade-off (ex: AIC, BIC, adjusted $R^2$)
        - Parameters: $\beta's$
            - Training - Once degree is chosen, there are well-defined values for the $\beta's$ that best fit the training data that can be algorithmically found by the computer


```
# Code from https://jakevdp.github.io/PythonDataScienceHandbook/05.03-hyperparameters-and-model-validation.html
# Thank you again Jake Vanderplas

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def make_data(N=30, err=0.8, rseed=1):
    # randomly sample the data
    rng = np.random.RandomState(rseed)
    X = rng.rand(N, 1) ** 2
    y = 10 - 1. / (X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X, y
    
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

def PolynomialRegression(degree=2, **kwargs):
    return make_pipeline(PolynomialFeatures(degree),
                         LinearRegression(**kwargs))

X, y = make_data()
xfit = np.linspace(-0.1, 1.0, 1000)[:, None]
model1 = PolynomialRegression(1).fit(X, y)
model20 = PolynomialRegression(20).fit(X, y)

fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

ax[0].scatter(X.ravel(), y, s=40)
ax[0].plot(xfit.ravel(), model1.predict(xfit), color='gray')
ax[0].axis([-0.1, 1.0, -2, 14])
ax[0].set_title('High-bias model: Underfits the data', size=14)

ax[1].scatter(X.ravel(), y, s=40)
ax[1].plot(xfit.ravel(), model20.predict(xfit), color='gray')
ax[1].axis([-0.1, 1.0, -2, 14])
ax[1].set_title('High-variance model: Overfits the data', size=14);
```

```
from ipywidgets import *
def update(degree=1):
    xfit = np.linspace(-0.1, 1.0, 1000)[:, None]
    model = PolynomialRegression(degree).fit(X, y)

    fig, ax = plt.subplots(1, 1, figsize=(16, 6))
    ax.scatter(X.ravel(), y, s=40)
    ax.plot(xfit.ravel(), model.predict(xfit), color='gray')
    ax.axis([-0.1, 1.0, -2, 14])
    plt.text(0.0, 12.0,f'$R^2$ = {model.score(X, y)}',fontsize=16 )

widget_degree = IntSlider(min=0, max=30, step=1, value=1, continuous_update=False)
interact(update, degree=widget_degree);
```

# Supervised Learning Project Workflow

data_science_flow.png

- Split Dataset - We split our observation into subsets for different purposes
    - *holdout set*
        - removed immediately
        - kept hidden until the end of model building
        - used to estimate generalization performance (how well model will perform on not-yet-seen samples)
        - not always done if sample size too small (but good practice)
    - *modeling set* - observations not in holdout set; further split into ...
        - *training set* - observations used to train the model

        - *validation / testing set* - observations used to evaluate the model
                
- Model scores
    - *training score* - model performance on training set
        - ignore this - its like giving a practice test to a class and then using THE SAME questions as the real test
    - *validation score* - model performance on validation set
        - pay attention - it is a reasonable measure of model performance
    - *generalization score* - model performance on holdout set
        - pay a lot of attention - it is the most rigorous measure of model's future performance on not-yet-seen samples


- *Cross-validation* - split modeling set into training & validation sets, train & test model, repeat.  Gives more robust estimate of model performance & helps improve quality of model selection (tuning) decisions.
 

# Weak/Strong and Lazy/Eager Learners

- Eager vs Lazy Learners - Training a model can take significant computation ...
    - An *eager* learner does most of the computations during model building and is therefore quick when applied to not-yet-seen samples (generalization)
    - A *lazy* learner must do a significant proportion of its computation when applied to not-yet-seen samples (generalization)

- [*Wisdom of Crowds*](https://en.wikipedia.org/wiki/Wisdom_of_the_crowd) - Often, a bunch of dummies do better than one smartie
    - Sir Francis Galton
    - booth at 1906 Plymouth County Fair
    - "Can you estimate the weight of this ox"
    - Lot's of wild guesses
    - Median within 1% of correct answer
    - Better than experts

- [*Ensemble Methods*](https://en.wikipedia.org/wiki/Ensemble_learning)
    - *weak learners* - Individual machine learners often do not perform well
    - *strong learner* - Combine a bunch of dummies → get a smartie
        - Random forests
        - Bagging
        - Boosting
        - AdaBoosting
        - Gradient Boosting
        - Stacking
